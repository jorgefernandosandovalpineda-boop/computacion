```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Generación de 700 registros
np.random.seed(42) # Para resultados consistentes
n_registros = 700

data = {
    "ID": range(1, n_registros + 1),
    "NOMBRE_VEHICULO": np.random.choice(
        ['Toyota Corolla', 'Ford F-150', 'Honda Civic', 'Nissan Sentra', 'VW Jetta', 'Hyundai Tucson'],
        n_registros
    ),
    "PRECIO": np.random.randint(15000, 55000, n_registros),
    "PLACAS_NUMERICAS": np.random.randint(1000, 9999, n_registros),
    "VENTAS_VEHICULO": np.random.randint(5, 100, n_registros),
    "VEHICULOS_EXISTENCIA": np.random.randint(0, 50, n_registros)
}

df = pd.DataFrame(data)

# Guardar a CSV
df.to_csv("datos_vehiculos_estadistica.csv", index=False)

# 2. ESTADÍSTICA DESCRIPTIVA BÁSICA
print("--- RESULTADOS ESTADÍSTICOS ---")

# Promedio de las ventas
promedio_ventas = df['VENTAS_VEHICULO'].mean()
print(f"Promedio de Ventas: {promedio_ventas:.2f}")

# Ventas Mínimo y Máximo
print(f"Ventas Mínimas: {df['VENTAS_VEHICULO'].min()}")
print(f"Ventas Máximas: {df['VENTAS_VEHICULO'].max()}")

# Precio Mínimo y Máximo
print(f"Precio Mínimo: ${df['PRECIO'].min()}")
print(f"Precio Máximo: ${df['PRECIO'].max()}")

# Desviación Estándar de Ventas y Precio
std_ventas = df['VENTAS_VEHICULO'].std()
std_precio = df['PRECIO'].std()
print(f"Desviación STD de Ventas: {std_ventas:.2f}")
print(f"Desviación STD de Precio: {std_precio:.2f}")

# 3. VISUALIZACIONES CON MATPLOTLIB

# Gráfico de barras: Precio promedio por vehículo
precio_promedio = df.groupby("NOMBRE_VEHICULO")["PRECIO"].mean()
plt.bar(precio_promedio.index, precio_promedio.values)
plt.title("Precio promedio por vehículo")
plt.xticks(rotation=45)
plt.ylabel("Precio ($)")
plt.show()

# Histograma: Distribución de ventas
plt.hist(df["VENTAS_VEHICULO"], bins=15, color="skyblue", edgecolor="black")
plt.title("Distribución de ventas de vehículos")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.show()
```
