```python
import pandas as pd
import numpy as np

# 1. Configuración de parámetros
n_registros = 50000
nombres = ["Ana", "Juan", "Maria", "Luis", "Elena", "Carlos", "Sofia", "Diego"]
apellidos = ["Garcia", "Lopez", "Perez", "Rodriguez", "Martinez", "Sanchez"]
paises = ["Mexico", "España", "Colombia", "Argentina", "Chile", "Peru"]

# 2. Creación de los datos aleatorios
data = {
    'id': np.arange(1, n_registros + 1),
    'nombre': np.random.choice(nombres, n_registros),
    'apellido': np.random.choice(apellidos, n_registros),
    'genero': np.random.choice(['F', 'M'], n_registros),
    'pais': np.random.choice(paises, n_registros),
    'estado': np.random.choice(['Soltero(a)', 'Casado(a)'], n_registros),
    'salario': np.random.randint(1500, 10001, n_registros),
    'edad': np.random.randint(18, 81, n_registros),
    # Para la fecha, generamos días aleatorios del año 2026 (365 días)
    'registro': pd.to_datetime('2026-01-01') + pd.to_timedelta(np.random.randint(0, 365, n_registros), unit='D')
}

df = pd.DataFrame(data)

# --- Resolución de problemas ---

# A. Despliegue una vista previa del DF
print("### Vista previa del DataFrame ###")
print(df.head())
print("\n" + "-"*50 + "\n")

# B. Muestre la estructura del DF (columnas y tipos de datos)
print("### Estructura del DF ###")
print(df.info())
print("\n" + "-"*50 + "\n")

# C. Muestre un análisis estadístico de las columnas numéricas
print("### Análisis estadístico ###")
print(df.describe())
print("\n" + "-"*50 + "\n")

# D. Muestre cuántos hombres y mujeres hay registrados
print("### Conteo de género ###")
print(df['genero'].value_counts())
print("\n" + "-"*50 + "\n")

# E. Muestre el conteo de hombres y mujeres por país
print("### Conteo de género por país ###")
conteo_pais_genero = df.groupby(['pais', 'genero']).size().unstack()
print(conteo_pais_genero)
```
